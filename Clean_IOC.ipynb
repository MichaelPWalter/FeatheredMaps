```python
import pandas as pd

# Replace path with the path to your Excel file
file_path = 'C:\\Users\\Walter\\Documents\\GitHub\\FeatheredMaps\\IOC_Names_File_Plus-14.2.xlsx'
df = pd.read_excel(file_path)

# Display the DataFrame
print(df.head())
```

```python
df.rename(columns={'SP; Extinct (†)': 'Extinct'}, inplace=True)
df['Extinct'] = df['Extinct'].replace('†', 1)

# Drop columns by ind
df.to_csv(f"{file_path.split("\\")[-1].split(".")[0]}.csv")
```

```python

# Loop through the rows
for index, row in df.iterrows():
    if row['Rank'] == 'Species':
        print("Species")
        # For species, abbreviate the scientific name
        species_name = row['Scientific Name'] # Get the full species name (eg. 'Struthio camelus')
        # Form the abbreviation
        parts = species_name.split(' ') # Separate the full scientific name (eg. 'Struthio', 'camelus')
        abbreviated_name = '. '.join([part[0] for part in parts]) + '.' # Take the first letter in each string and join them (eg. 'S. c.')
        print(abbreviated_name)

    elif row['Rank'] == 'ssp':
        print("ssp")
        # For subspecies, replace the abbreviation with the full species name
        subspecies_name = row['Scientific Name'] # Get the subspecies name (eg. '    S. c. syriacus')
        subspecies_name = subspecies_name.strip().split(' ') # Split the subspecies name (eg. 'S.', 'c.', 'syriacus')
        abbreviated_species = ' '.join(subspecies_name[:2])  # Join the abbreviated part (e.g., 'S. c.')
        print(abbreviated_species)
        if abbreviated_name == abbreviated_species: # Compare the abbreveated_name from the Species to the ssp
            fullsubspecies_name = ' '.join([species_name,subspecies_name[-1]]) # Join species and subspecies str (eg. 'Struthio camelus syriacus')
            print(fullsubspecies_name)
            df.at[index, 'Scientific Name'] = fullsubspecies_name # Overwrite the scientific Name of the subspecies
    else:
        print("neither")

df.to_csv(f"Cleaned_{file_path.split("\\")[-1].split(".")[0]}.csv")

```

```python
# Add new columns with default values (e.g., empty strings)
df['order'] = ''  
df['family'] = ''
df['genus'] = ''
df['species'] = ''

# Loop through the rows
for index, row in df.iterrows():
    if row['Rank'] == 'ORDER':
        print("ORDER")
        # For species, abbreviate the scientific name
        Order = row['Scientific Name'] # Get the full species name (eg. 'Struthio camelus')
        # Form the abbreviation
        # Remove the word 'ORDER'
        Order = Order.replace("ORDER", "").strip().lower().capitalize()
        print(Order)

    elif row['Rank'] == 'Family':
        print("Family")
        # For subspecies, replace the abbreviation with the full species name
        Family = row['Scientific Name']
        Family = Family.replace("Family", "").strip().lower().capitalize()
        df.at[index, 'order'] = Order
        print(Family)

    elif row['Rank'] == 'Genus':
        print("Genus")
        # For subspecies, replace the abbreviation with the full species name
        Genus = row['Scientific Name'].strip().lower().capitalize()
        df.at[index, 'order'] = Order
        df.at[index, 'family'] = Family
        print(Genus)   

    elif row['Rank'] == 'Species':
        print("Species")
        # For species, abbreviate the scientific name
        Species = row['Scientific Name'] # Get the full species name (eg. 'Struthio camelus')
        df.at[index, 'order'] = Order
        df.at[index, 'family'] = Family
        df.at[index, 'genus'] = Genus
        print(Species)

    elif row['Rank'] == 'ssp':
        print("Subspecies")
        # For species, abbreviate the scientific name
        Subspecies = row['Scientific Name'] # Get the full species name (eg. 'Struthio camelus')
        df.at[index, 'order'] = Order
        df.at[index, 'family'] = Family
        df.at[index, 'genus'] = Genus
        df.at[index, 'species'] = Species
        print(Subspecies)
        
    else:
        print("neither")



df.to_csv(f"Cleaned_{file_path.split("\\")[-1].split(".")[0]}.csv")
```
