```python
import pandas as pd

file_path = "C:\\Users\\Walter\\Documents\\GitHub\\FeatheredMaps\\Clements-v2024-October-2024-rev.csv"
df = pd.read_csv(file_path)

```

```python
# Add new columns with default values (e.g., empty strings)
df.rename(columns={'family': 'family_english_Clements', 'order': 'order_english_Clements'}, inplace=True)
 
df['family'] = ''
df['genus'] = ''
df['species'] = ''

# Loop through the rows
for index, row in df.iterrows():

    if row['category'] == 'family':
        print("Family")
        Family = row['scientific name']
        Family = Family.strip()
        print(Family)


    elif row['category'] == 'species':
        print("Species")
        Species = row['scientific name'] # Get the full species name (eg. 'Struthio camelus')
        Genus = Species.split(" ")[0]
        df.at[index, 'family'] = Family
        df.at[index, 'genus'] = Genus
        print(Species)

    elif row['category'] == 'subspecies':
        print("Subspecies")
        Subspecies = row['scientific name'] # Get the full species name (eg. 'Struthio camelus')
        df.at[index, 'family'] = Family
        df.at[index, 'genus'] = Genus
        df.at[index, 'species'] = Species
        print(Subspecies)

    elif row['category'] == 'group (monotypic)':
        print("Group (Monotypic)")
        Subspecies = row['scientific name'] # Get the full species name (eg. 'Struthio camelus')
        df.at[index, 'family'] = Family
        df.at[index, 'genus'] = Genus
        df.at[index, 'species'] = Species
        print(Subspecies)
    
    elif row['category'] == 'form':
        print("Form")
        Subspecies = row['scientific name'] # Get the full species name (eg. 'Struthio camelus')
        df.at[index, 'family'] = Family
        df.at[index, 'genus'] = Genus
        df.at[index, 'species'] = Species
        print(Subspecies)
        
    else:
        print("neither")

df.to_csv(f"Cleaned_{file_path.split("\\")[-1].split(".")[0]}.csv")
```
